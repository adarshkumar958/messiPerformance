```python
def get_half(x):
    if '+' in x:
        return 'extra_time'
    elif int(x)<46:
        return 'first_half'
    else:
        return 'second_half'

messi['goal_halves']=messi.minute.apply(lambda x:get_half(x))
messi.head()

```

```python
def transform(y):
    if '+' in y:
        
        sum1= y.split('+')
        return int(sum1[0])+int(sum1[1])
        
    else:
        return int(y)
    
messi.minute=messi.minute.apply(lambda y:transform(y))
messi.minute.astype(int)
messi.head()
```
